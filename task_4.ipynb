```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
size = [500, 1000]
X_range = [(0,1000), (100,600)]
Y_range = [(0,1000), (10,500)]
N_claster = 3
Colors = np.array(["r", 'b', 'g', 'yellow', 'black', 'orange'])
```

```python
X, Y = [],[]
for i in range(len(size)):
    X = [*X, *np.random.randint(X_range[i][0], X_range[i][1], size[i])]
    Y = [*Y, *np.random.randint(Y_range[0][0], Y_range[i][1], size[i])]
```

```python
X = np.random.randint(X_range[0], X_range[1], size)
Y = np.random.randint(Y_range[0], Y_range[1], size)
```

```python
plt.scatter(X, Y, s=1, c='black')
```

```python
clusterizator = KMeans(N_claster)
```

```python
prediction = clusterizator.fit_predict([(X[i], Y[i]) for i in range(sum(size))])
```

```python
plt.scatter(X, Y, s=1, c=[Colors[val] for val in prediction])
```
